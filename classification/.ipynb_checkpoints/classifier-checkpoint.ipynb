```python

import numpy as np 
import wfdb

def load_raw_data(df, sampling_rate, path):
    if sampling_rate == 100:
        data = [wfdb.rdsamp(path+f) for f in df.filename_lr]
    else:
        data = [wfdb.rdsamp(path+f) for f in df.filename_hr]
    data = np.array([signal.T for signal, meta in data])
    return data
```

```python
from sklearn.metrics import confusion_matrix, roc_auc_score, classification_report, accuracy_score, f1_score, recall_score, precision_score


def validate_model(y_trues, y_preds, resolution = "LR"):

    acc = accuracy_score(y_trues, y_preds)
    f1 = f1_score(y_trues, y_preds, average = 'macro')
    specificity = recall_score(y_trues, y_preds, average = 'macro', pos_label=0)
    sensitivity = recall_score(y_trues, y_preds, average = 'macro')
    precision =  precision_score(y_trues, y_preds, average = 'macro')
    aucs = roc_auc_score_multiclass(y_trues, y_preds, average = "macro")
    aucs = list(aucs.values())
    auc = np.mean(aucs)


    cr = classification_report(y_trues, y_preds)
    cm = confusion_matrix(y_trues, y_preds)
    metrics = {
        "accuracy": acc,
        "f1": f1,
        "cm": cm,
        "sensitivity": sensitivity,
        "specificity": specificity,
        "auc_score": auc,
        "precision": precision,
        "report": cr,
        "predictions": y_preds,
    }

    return metrics, "Accuracy: {} \n F1 score: {} \n Sensitivity: {} \n Specificity: {} \n ROC AUC score: {} \n Confusion Matrix: \n {} \n Classification Report: \n {} \n".format(acc, f1, sensitivity, specificity, auc, cm, cr)
```

```python
def plot_cm(classes, cm, resolution = "LR"):

    import matplotlib.pyplot as plt
    import seaborn as sns
    import pandas as pd
    import numpy as np
    import itertools

    sns.set(style='white')
    fig, ax = plt.subplots(figsize=(12, 8))
    plt.imshow(cm, cmap = "Blues")
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    ax.tick_params(labelsize=20, length=0)

    ax.set_title('Confusion Matrix CNN classifier ECG {} resolution'.format(resolution), size=24, pad=20)
    ax.set_xlabel('Predicted', size=20)
    ax.set_ylabel('Target', size=20)

    nclasses = len(classes)
    ticks = np.arange(0, nclasses, 1)
    ax.set_xticks(ticks)
    ax.set_yticks(ticks)
    ax.set_xticklabels(classes)
    ax.set_yticklabels(classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center", size = 20,
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.show()
```

```python
def roc_auc_score_multiclass(actual_class, pred_class, average = "macro"):

    #creating a set of all the unique classes using the actual class list
    unique_class = np.unique(actual_class)
    roc_auc_dict = {}
    for per_class in unique_class:

        #print(per_class)
        #creating a list of all the classes except the current class
        other_class = [x for x in unique_class if x != per_class]

        #marking the current class as 1 and all other classes as 0
        new_actual_class = [0 if x in other_class else 1 for x in actual_class]
        new_pred_class = [0 if x in other_class else 1 for x in pred_class]

        #using the sklearn metrics method to calculate the roc_auc_score
        roc_auc = roc_auc_score(new_actual_class, new_pred_class, average = average)
        roc_auc_dict[per_class] = roc_auc

    return roc_auc_dict
```

```python
import scipy 

def resample_signal(data, fs=50, nchs = 12, fin = 100):

    data = data.cpu().detach().numpy()

    secs = data.shape[-1]/(fin)
    size = int(fs*secs)

    resampled_data = []
    for ch in range(nchs):
        temp = scipy.signal.resample(x=data[ch, :], num=size)
        resampled_data.append(torch.from_numpy(temp))

    resampled_data = torch.stack(resampled_data)

    return resampled_data
```

```python
import numpy as np

def superresolute(lr_raw):
    
    sr_raw = []
    n = len(lr_raw)
    for i, x_lr in enumerate(lr_raw):

         if i in np.arange(1, n, 1000):
            print(i+1, "/", n, end="\r")
            torch.cuda.empty_cache()
            gc.collect()
         x_lr = torch.from_numpy(x_lr)
         signal_filt = myfilter(0.05, x_lr)
         fs = 50
         signal_filt = resample_signal(signal_filt, fs=fs, fin = 100)
         windows = sliding_window(signal_filt, size = fs*5, stride = fs*5)
         windows_sr = []
         for window in windows:
             window = window.cpu().detach().numpy()
             window = add_noise_single(window, fs = fs)
             window = torch.tensor(window).float().to(device)
             pred = model1d_sr([window, None, None, None])
             window_sr = pred[1].cpu()#.to(device)
             window_sr = resample_signal(window_sr, fs=100, fin = 500)
             windows_sr.append(window_sr)
             del window
             del pred

         sr_signal = torch.cat(windows_sr, axis = 1)
         sr_raw.append(sr_signal)
         del windows
         del window_sr

    return sr_raw
```

```python
import random

def add_rsp(signal, fs = 100):

    nchs = 12
    duration = 5

    while True:
        rr_all = np.arange(10, 20, 2)
        rr = random.choice(rr_all)
        #print(rr)
        resp = nk.rsp_simulate(length=fs*duration, sampling_rate = fs, respiratory_rate=rr, method="breathmetrics")

        temp = [torch.from_numpy(resp) for i in range(nchs)]
        resp = torch.from_numpy(np.array(temp))

        if signal.shape == resp.shape:
            break
    noisy_signal = signal.cpu() + resp

    return noisy_signal
```

```python
def add_eda(signal, fs = 100):

    nchs = 12
    duration = 5

    edan = random.choice(np.arange(1, 2))
    eda = nk.eda_simulate(duration = duration, length=fs*duration, sampling_rate = fs, scr_number=edan, drift=-0.5)

    temp = [torch.from_numpy(eda) for i in range(nchs)]
    eda = torch.from_numpy(np.array(temp))
    noisy_signal = signal.cpu() + eda
    return noisy_signal
```

```python
def add_emg(signal, fs = 100):

    nchs = 12
    duration = 5

    bn = random.choice(np.arange(1, 2))
    emg = nk.emg_simulate(length=fs*duration, sampling_rate = fs, burst_number=bn, burst_duration=1)

    temp = [torch.from_numpy(emg) for i in range(nchs)]
    emg = torch.from_numpy(np.array(temp))
    noisy_signal = signal.cpu() + emg
    return noisy_signal
```

```python
def add_noise_single(signal, fs = 100):
    
    noise_functions = {
    "Respiration Noise": add_rsp,
    "EMG artifact": add_emg,
    "EDA artifact": add_eda
    }
    
    idxs =  [0, 1, 2, 3, 4] 
    idx = random.choice(idxs)
    signal = torch.from_numpy(signal)
    if idx >= 3:
        
            key = "Normal"
            signal_filt = myfilter(0.05, signal)

    else:
        
            keys = list(noise_functions.keys())
            key = keys[idx]
            noise_function = noise_functions[key]
            noisy_signal = noise_function(signal, fs = fs) 
            signal_filt = myfilter(0.05, noisy_signal)
            signal_filt = torch.squeeze(signal_filt, dim = 0)
            signal_filt = torch.squeeze(signal_filt, dim = 0)
        
    return signal_filt
```

```python
import random 
import neurokit2 as nk 

def add_noise(lr_raw, fs = 100):
    lr_noisy = []
    for signal in lr_raw:         
         fs = 100
         windows = sliding_window(signal, size = fs*5, stride = fs*5)
         noisy_windows = []
         for window in windows:
            noisy_window = add_noise_single(window,  fs = 100)
            noisy_windows.append(noisy_window)
             
         signal_filt = torch.cat(noisy_windows, axis = 1)
         lr_noisy.append(signal_filt)
    return lr_noisy
```

```python
from scipy.signal import butter, lfilter, iirnotch

def bandpass(lowcut, highcut, order=5, fs = 100):

    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_highpass(highcut, fs=100, order=5):
    nyq = 0.5 * fs
    high = highcut / nyq
    b, a = butter(order, high, btype='high')#, analog = True)
    return b, a

def notch_filter(cutoff, q, fs=100):

    nyq = 0.5*fs
    freq = cutoff/nyq
    b, a = iirnotch(freq, q)
    return b, a

def myfilter(lowcut, data, powerline=None, fs = 100):

    nchs = 12
    filtered_data = torch.zeros_like(data)
    for ch in range(nchs):
        ch_data = data[ch, :]
        b, a = butter_highpass(lowcut, fs = fs)
        x = lfilter(b, a, ch_data)
        if powerline is not None:
            f, e = notch_filter(powerline, 30, fs = fs)
            x = lfilter(f, e, x)
        filtered_data[ch, :] = torch.from_numpy(x)
    return filtered_data
```

```python
import math
def sliding_window(signal, size=250, stride=250):
    windows = []
    sig_len = signal.shape[-1]
    win_num = math.ceil((sig_len - size) / stride) + 1
    for i in range(win_num):
        offset = i * stride
        windows.append((signal[:, offset : offset + size]))
    return windows
```

```python
from sklearn.preprocessing import StandardScaler

def apply_scaler(inputs, scaler):
    """Applies standardization to each individual ECG signal.

    Parameters
    ----------
    inputs: np.array
        Array of ECG signals.
    scaler: StandardScaler
        Standard scaler object.

    Returns
    -------
    np.array
        Array of standardized ECG signals.

    """

    temp = []
    for x in inputs:
        x_shape = x.shape
        temp.append(scaler.transform(x.flatten()[:, np.newaxis]).reshape(x_shape))
    temp = np.array(temp)
    return temp
```

```python
resolutions = ["original", "lr + noise", "sr"] 
classes = np.unique(y_raw)
print(classes)

n = len(y_raw)

model.to(device)
model.eval()

ypreds_dict = {}

for resolution in resolutions:
    print(resolution)
    y_preds = []
    if resolution == "original":
        xtest = lr_raw
    elif resolution == "lr + noise":
        xtest = add_noise(lr_raw, fs = 50)
    else:
        xtest = superresolute(lr_raw)

    # Standardization
    scaler = StandardScaler()
    scaler.fit(np.vstack(xtest).flatten()[:, np.newaxis].astype(float))
    xtest = apply_scaler(xtest, scaler)
    
    model.eval()
    for i, x in enumerate(xtest):
        if i in np.arange(1, len(xtest), 100):
            print(i+1, "/", len(xtest), end = "\r")
        
        x = torch.tensor(x).float()
        x = torch.unsqueeze(x, dim = 0).to(device)
        pred = model(x)
        y_preds.append(torch.argmax(pred).cpu())
        del x
        del pred 
    ypreds_dict[resolution] = y_preds
```
