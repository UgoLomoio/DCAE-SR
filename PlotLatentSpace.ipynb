```python
import gc
from matplotlib.pyplot import rcParams
import os
import warnings
import matplotlib.pyplot as plt

gc.collect()
with torch.no_grad():
    torch.cuda.empty_cache()

warnings.simplefilter(action='ignore', category=FutureWarning)
rcParams['font.weight'] = 'bold'
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
```

```python
import gc
gc.collect()
torch.cuda.empty_cache()
```

```python
map_superclass_rev = {'CD': 0, 'HYP': 1, 'MI': 2, 'NORM': 3, 'STTC': 4}
map_superclass_rev
```

```python
import torch

labels = map_superclass_rev.keys()
dict_windows_test_hr = {}
dict_windows_test_lr = {}
path = os.getcwd()+os.sep+"pt_data"+os.sep
device = "cuda" if torch.cuda.is_available() else "cpu"

for i, label in enumerate(labels):

    print("\r", i+1, "/", len(labels), label, end = "")
    filename = path + "test_{}_hr.pt".format(label)
    hr_data = torch.load(filename, map_location=torch.device(device))
    dict_windows_test_hr[label] = hr_data

    filename = path + "test_{}_noisy.pt".format(label)
    lr_data = torch.load(filename, map_location=torch.device(device))
    dict_windows_test_lr[label] = lr_data
```

```python
import os

cwd = os.getcwd()
model1d = torch.load(cwd+"/models/model_2_denoising_lrhr_v2.pt", map_location=torch.device(device))
model1d
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scale(ecg, nchs = 12):

  if ecg.ndim == 3:
    ecg = torch.squeeze(ecg, dim = 0)

  tf = MinMaxScaler((-1, 1))#StandardScaler()
  scaled = tf.fit_transform(ecg.cpu().detach().numpy())
  scaled = torch.from_numpy(scaled)
  return scaled
```

```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA

colors = ["red", "orange", "blue", "black", "purple"]

# here we are creating sub plots
figure, ax = plt.subplots(figsize=(10, 8))

for i, label in enumerate(labels):

  print("\r", i+1, "/", len(labels), end = "")
  if label != "MI":
    pass
    #continue

  X_test_lr = dict_windows_test_lr[label]

  points = []

  for x in X_test_lr:
    # We only need to encode the data
    proj = model1d.encoder(x)
    proj = proj.flatten()
    points.extend([proj.detach().cpu().numpy()])
    del x

  points = np.array(points)
  points = TSNE(n_components=2, learning_rate="auto", init='random', perplexity=25.0).fit_transform(points)
  #points = PCA(n_components=2, svd_solver='randomized').fit_transform(points)
  n = len(X_test_lr)
  label_idcs = [i for j in range(n)]

  plt.scatter(x=points[:, 0], y=points[:, 1], s=2.0, c=colors[i], label = label)

lgnd = plt.legend()

for handle in lgnd.legend_handles:
    handle.set_sizes([28.0])

plt.legend()
```
